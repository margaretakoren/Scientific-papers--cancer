```python
new_columns = [
    'Institution', 'Address', 'Authors', 'Journal', 'Impact Factor', 
    'Title', 'No. of citations', 'Year', 'Type of study'
]
new_df = pd.DataFrame(columns=new_columns)
new_df 
```

```python
column_mapping = {
    'Title': 'Title',
    'Journal': 'Journal Abbreviation',
    'Authors': 'Author',
    'Year': 'Publication Year'
}
```

```python
for new_col, old_col in column_mapping.items():
    if old_col in df.columns:
        new_df[new_col] = df[old_col]
    else:
        new_df[new_col] = np.nan

new_df.head()
```

```python
Entrez.email = "margaretakoren03@gmail.com"


def fetch_pmid(title):
    handle = Entrez.esearch(db="pubmed", term=title, retmax=1)
    record = Entrez.read(handle)
    handle.close()
    id_list = record.get("IdList", [])
    if id_list:
        return id_list[0]
    return None

df['PMID'] = df['Title'].apply(fetch_pmid)

df.head()
```

```python
def fetch_pubmed_details(pmid):
    try:
        handle = Entrez.esummary(db="pubmed", id=pmid, retmode="xml")
        record = Entrez.read(handle)
        handle.close()
        summary = record[0]
        return summary
    except Exception as e:
        print(f"Error fetching details for PMID '{pmid}': {e}")
        return None

df['PubMed_Details'] = df['PMID'].apply(lambda pmid: fetch_pubmed_details(pmid) if pd.notna(pmid) else None)


df.head()
```

```python
def fetch_croatian_affiliations(pmid):
    if pd.isna(pmid) or pmid is None:
        return None
    try:
        handle = Entrez.efetch(db="pubmed", id=pmid, retmode="xml")
        records = Entrez.read(handle)
        handle.close()
        
        affiliations = set()
        article = records['PubmedArticle'][0]['MedlineCitation']['Article']
        if 'AuthorList' in article:
            for author in article['AuthorList']:
                if 'AffiliationInfo' in author:
                    for affil in author['AffiliationInfo']:
                        affiliation_text = affil['Affiliation']
                        if 'Croatia' in affiliation_text:
                            affiliations.add(affiliation_text)
        return "; ".join(affiliations) if affiliations else None
    except Exception as e:
        print(f"Error fetching affiliations for PMID '{pmid}': {e}")
        return None

affiliation_data = []
for pmid in df['PMID']:
    if pmid: 
        affiliations = fetch_croatian_affiliations(pmid)
        if affiliations: 
            affiliation_data.append({'PMID': pmid, 'Affiliation': affiliations})

croatian_affiliations_df = pd.DataFrame(affiliation_data)
croatian_affiliations_df.sort_values(by='PMID', inplace=True)
croatian_affiliations_df.head()
```
